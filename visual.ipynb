```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import numpy as np

# Load the JSON data
file_path = 'f_son_heung_min_match_stats.json'  #      
with open(file_path, 'r') as file:
    data = json.load(file)

# Create a DataFrame
df = pd.DataFrame.from_dict(data, orient='index')

# Convert 'date' column to datetime format
df.index = pd.to_datetime(df.index)
df.reset_index(inplace=True)
df.rename(columns={'index': 'date'}, inplace=True)

# Extract year from date for season analysis
df['season'] = df['date'].dt.year

# Convert 'goals' and 'assists' columns to numeric (some might be empty strings)
df['goals'] = pd.to_numeric(df['goals'], errors='coerce')
df['assists'] = pd.to_numeric(df['assists'], errors='coerce')

# Replace NaN values with 0
df = df.assign(goals=df['goals'].fillna(0))
df = df.assign(assists=df['assists'].fillna(0))

# Add a column for total contributions (goals + assists)
df['total_contributions'] = df['goals'] + df['assists']

# Convert 'minutes' column to numeric, handle errors by coercion
df['minutes'] = pd.to_numeric(df['minutes'], errors='coerce')

# Add a time_bins column
time_bins = [0, 15, 30, 45, 60, 75, 90]
labels = ['0-15', '16-30', '31-45', '46-60', '61-75', '76-90']
df['time_bin'] = pd.cut(df['minutes'], bins=time_bins, labels=labels, right=False)

# Filter rows with valid minutes values
df = df[~df['minutes'].isnull()]

# Group by time_bin and calculate sums for goals and assists separately
goals_per_time_bin = df.groupby('time_bin', observed=True)['goals'].sum().reset_index()
assists_per_time_bin = df.groupby('time_bin', observed=True)['assists'].sum().reset_index()

# Calculate the number of matches per time_bin where Son had contributions
matches_with_contributions = df.groupby('time_bin', observed=True)['date'].count().reset_index()
matches_with_contributions.columns = ['time_bin', 'matches']

# Merge the data
goals_per_time_bin = pd.merge(goals_per_time_bin, matches_with_contributions, on='time_bin')
assists_per_time_bin = pd.merge(assists_per_time_bin, matches_with_contributions, on='time_bin')

# Calculate win probability
df['win'] = df['result'].apply(lambda x: 1 if 'Win' in x else 0)
win_probability_per_time_bin = df.groupby('time_bin', observed=True)['win'].mean().reset_index()
win_probability_per_time_bin.columns = ['time_bin', 'win_probability']

# Set the style of seaborn
sns.set(style="whitegrid")

#    
plt.figure(figsize=(18, 15))
sns.barplot(x='season', y='total_contributions', data=df.groupby('season')['total_contributions'].sum().reset_index(), hue='season', palette="Blues_d", dodge=False)
plt.xlabel('Season')
plt.ylabel('Total Contributions (Goals + Assists)')
plt.title('Total Contributions per Season by Son Heung-Min')
plt.xticks(rotation=90)
plt.grid(True)
plt.legend([], [], frameon=False)
plt.savefig('total_contributions_per_season.png')
plt.show()

#     
fig, ax1 = plt.subplots(figsize=(18, 15))

#   
color = 'tab:blue'
ax1.set_xlabel('Time Bin')
ax1.set_ylabel('Goals', color=color)
sns.barplot(x='time_bin', y='goals', data=goals_per_time_bin, hue='time_bin', palette="Blues_d", dodge=False, ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Goals and Win Probability per Time Bin by Son Heung-Min')
plt.legend([], [], frameon=False)

#   
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Win Probability', color=color)
sns.lineplot(x='time_bin', y='win_probability', data=win_probability_per_time_bin, sort=False, color=color, marker='o', ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.grid(True)
plt.savefig('goals_and_win_probability_per_time_bin.png')
plt.show()

#     
fig, ax1 = plt.subplots(figsize=(18, 15))

#   
color = 'tab:green'
ax1.set_xlabel('Time Bin')
ax1.set_ylabel('Assists', color=color)
sns.barplot(x='time_bin', y='assists', data=assists_per_time_bin, hue='time_bin', palette="Greens_d", dodge=False, ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Assists and Win Probability per Time Bin by Son Heung-Min')
plt.legend([], [], frameon=False)

#   
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Win Probability', color=color)
sns.lineplot(x='time_bin', y='win_probability', data=win_probability_per_time_bin, sort=False, color=color, marker='o', ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.grid(True)
plt.savefig('assists_and_win_probability_per_time_bin.png')
plt.show()

#  +   
fig, ax1 = plt.subplots(figsize=(18, 15))

#  + 
color = 'tab:blue'
ax1.set_xlabel('Time Bin')
ax1.set_ylabel('Total Contributions', color=color)
sns.barplot(x='time_bin', y='total_contributions', data=df.groupby('time_bin', observed=True)['total_contributions'].sum().reset_index(), hue='time_bin', palette="Blues_d", dodge=False, ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Total Contributions and Win Probability per Time Bin by Son Heung-Min')
plt.legend([], [], frameon=False)

#   
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Win Probability', color=color)
sns.lineplot(x='time_bin', y='win_probability', data=win_probability_per_time_bin, sort=False, color=color, marker='o', ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.grid(True)
plt.savefig('contributions_and_win_probability_per_time_bin.png')
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load the JSON data
file_path = 'match_logs.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Transform the data into a DataFrame
records = []
for date, events in data.items():
    for event in events:
        records.append({
            'date': date,
            'time': event.get('time'),
            'type': event.get('type')
        })

df = pd.DataFrame(records)

# Convert 'date' to datetime format and extract year
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

# Group by event type and year
event_type_year = df.groupby(['year', 'type']).size().unstack().fillna(0)

# Plot the distribution of event types over the years
fig, ax = plt.subplots(figsize=(12, 8))
event_type_year.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Distribution of Event Types Over the Years')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Events')
ax.legend(title='Event Type')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot as an image
plt.savefig('event_distribution_over_years.png')

# Show the plot
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load the match logs from the JSON file
with open("match_logs.json", "r") as file:
    match_logs = json.load(file)

# Extract goal and assist events with their respective times
events = []

for date, actions in match_logs.items():
    for action in actions:
        if action['type'] in ['goal', 'assist']:
            minute_str = action['time'].split('')[0].strip()
            try:
                minute = int(minute_str)
            except ValueError:
                minute = 90  # For cases like "90+2"
            events.append({"date": date, "minute": minute, "type": action['type']})

# Convert to DataFrame
df = pd.DataFrame(events)

# Define the time bins
time_bins = [0, 15, 30, 45, 60, 75, 90]
labels = ['0-15', '16-30', '31-45', '46-60', '61-75', '76-90']

# Create a column for the time bins
df['time_bin'] = pd.cut(df['minute'], bins=time_bins, labels=labels, right=False)

# Count the number of goals and assists in each time bin
bin_counts = df.groupby(['time_bin', 'type'], observed=False).size().unstack(fill_value=0)

# Plot the data
bin_counts.plot(kind='bar', stacked=True)
plt.title("Son Heung-min's Goals and Assists by Time Interval")
plt.xlabel("Time Interval (minutes)")
plt.ylabel("Number of Events")
plt.legend(title="Event Type")
plt.xticks(rotation=0)
plt.tight_layout()

# Save the plot as an image
plt.savefig("son_heung_min_goals_assists.png")

# Show the plot
plt.show()
```
